```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read the school data file and store it in a Pandas DataFrame.
#The read_csv() function makes it easier for us by converting the CSV file to a DataFrame.
#.head() will show the first five rows
#adding a number in the () will show you that many lines
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head(10)
```

```python
# Read the student data file and store it in a Pandas DataFrame.
#The read_csv() function makes it easier for us by converting the CSV file to a DataFrame.
#.tail will show you the last five rwos
#adding a number in the () will show you that many lines
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.tail(10)
```
